```python
import matplotlib.pyplot as plt
```

Harga Mahal

```python
x = []
y = []

for i in range(5000, 55000):
    x.append(i)
    if i > 25001:
        y.append(1)
    else:
        y.append(0)
plt.plot(x, y)
```

Harga Sedang


```python
x1 = []
y1 = []

for i in range(5000, 50000):
    x1.append(i)
    if i > 15001 and i <25000:
        y1.append(1)
    else:
        y1.append(0)
plt.plot(x1, y1)
```

Harga Murah

```python
x2 = []
y2 = []

for i in range(5000, 50000):
    x2.append(i)
    if i < 15000:
        y2.append(1)
    else:
        y2.append(0)
plt.plot(x2, y2)
```

```python
plt.plot(x, y, x1, y1, x2, y2)
```

Porsi Sedikit


```python
x3 = []
y3 = []

for i in range (100, 1100):
    x3.append(i)
    if i < 250:
        y3.append(1)
    else:
        y3.append(0)
plt.plot(x3,y3)
```

Porsi Banyak

```python
x4 = []
y4 = []

for i in range (100, 1100):
    x4.append(i)
    if i > 250 and i < 500:
        y4.append(1)
    else:
        y4.append(0)
plt.plot(x4,y4)
```
